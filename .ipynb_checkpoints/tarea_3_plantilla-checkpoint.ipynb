6. Remove non-english apps from both Android and Apple datasets. You should write a function that takes a string, if there's more than three characters that fall outside the ASCII range (0 - 127), then the function should return False (identify the string as non-English), otherwise it should return True. Hint: use the [```ord()```  function](https://www.w3schools.com/python/ref_func_ord.asp)

````python
# For instance
correct_english_name = "Instachat üòú"  # 1 character outside the ASCII range (üòú)
incorrect_english_name = "Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠" # 13 character outside the ASCII range (ie Êí≠)
````

````python
print(ord("I"))
print(ord("t"))
print(ord("-"))
print(ord("üòú"))
print(ord("Áà±"))
````

````python
# Solution
def is_english(x):
  for ch in x:
    if ord(ch) > 127:
      return False
  return True

for app in google.App:
    if is_english(app) is False:
      google.drop(google[google.App == app].index, inplace=True)

google
````

 8. Build a frequency table for the ```prime_genre``` column of the App Store dataset, and the ```Genres``` and ```Category``` columns of the Google Play data set. What are the most common genres for each market? You must develop and deploy your function

````python
# Solution
import pandas as pd

def generar_tabla_frecuencias(data, columna):
    tabla_frecuencias = data[columna].value_counts().reset_index()
    tabla_frecuencias.columns = [columna, 'Frecuencia']
    return tabla_frecuencias

tabla_app_store = generar_tabla_frecuencias(apple, 'prime_genre')

tabla_google_play_genres = generar_tabla_frecuencias(google, 'Genres')

tabla_google_play_category = generar_tabla_frecuencias(google, 'Category')

print("Tabla de frecuencias para App Store:")
print(tabla_app_store)

print("\nTabla de frecuencias para Google Play - G√©neros:")
print(tabla_google_play_genres)

print("\nTabla de frecuencias para Google Play - Categor√≠as:")
print(tabla_google_play_category)
````

 9. One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre. Based on that exercise, give one app profile recommendation for Google Play.
 
    For the Google Play data set, we can find the required information in the ```Installs``` column. Then, compute the average number of installs by genre. Hint 1: consider only unique values of the ```Category``` column. Hint 2: If an app has, for instance, ```1,000,000+``` assume it as ```1,000,000``` installs. Hint 3: Be careful with a skewed distribution

````python
# Solution

google['Installs'] = google['Installs'].str.replace(',', '')
google['Installs'] = google['Installs'].str.replace('+', '')
google['Installs'] = pd.to_numeric(google['Installs'], errors='coerce')

promedio_instalaciones_por_categoria = google.groupby('Category')['Installs'].mean().reset_index()

promedio_instalaciones_por_categoria = promedio_instalaciones_por_categoria.sort_values(by='Installs', ascending=False)

print("Recomendaci√≥n de Perfil de Aplicaci√≥n para Google Play:")
print(promedio_instalaciones_por_categoria[['Category', 'Installs']])
````

 10. The App Store dataset doesn¬¥t have the ```installs``` information. Then, you will construct the average rating from the ```rating_count_tot``` column per genre. Based on that exercise, give one app profile recommendation for the App Store

````python
# Solution

promedio_calificacion_por_genero = apple.groupby('prime_genre')['rating_count_tot'].mean().reset_index()

promedio_calificacion_por_genero = promedio_calificacion_por_genero.sort_values(by='rating_count_tot', ascending=False)

print("Recomendaci√≥n de Perfil de Aplicaci√≥n para la App Store:")
print(promedio_calificacion_por_genero[['prime_genre', 'rating_count_tot']])
````
